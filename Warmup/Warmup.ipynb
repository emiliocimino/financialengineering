```python
# Multivariate normal distribution.
#Mean 0, covmatrix = [[1, -0.5],[-0.5, 2]]

mu = np.zeros(2)
print(mu)
covmat = np.asarray([[1, -0.5],[-0.5, 2]])
print(covmat)

x, y= np.random.multivariate_normal(mean = mu, cov= covmat, size=1000).T
```

```python
def mean(x):
    s = 0
    for elem in x:
        s += elem
    return s/x.shape[0]
```

```python
def cov(x, y):
    mean_x = mean(x)
    mean_y = mean(y)
    num_sam = x.shape[0]
    cov = np.sum((x-mean_x).dot((y - mean_y).T))/num_sam
    return cov
```

```python
plt.scatter(x, y)
print(mean(x), mean(y))
print(cov(x,y))
print(np.asarray([[cov(x,x),cov(x,y)],[cov(y,x),cov(y,y)]]))
```
