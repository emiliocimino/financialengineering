# Synthetic 2 Stocks Portfolio

L'idea dietro questo notebook è simulare un portfolio con 2 stocks per capire un po' come funziona il tutto e fare un plot risk-return

```python
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt
np.random.seed(11111111)
```

```python
# Expected returns da distribuzione normale a media 0 e std dev 0.01
mean_return = np.random.randn(2)*0.01
mean_return
```

```python
# correlation tra le due variabili, normal distr. media 0 e varianza 0.01
rho = np.random.randn()*0.01
rho
```

```python
# Deviazioni standard. usiamo exponential per renderle positive
sigmas = np.exp(np.random.randn(2))
sigmas
```

```python
# Creazione matrice covarianza. Iniziamo dalle diagonali della matrice con np.diag.
# Le diaginali sono le varianze, ossia il quadrato delle std dev
cov = np.diag(sigmas**2)
cov
```

```python
# Ora andiamo di covarianze: calcoliamo la covarianza usando la formula cov(x,y) = corr(x,y)*sigma_x*sigma_y
sigma12 = rho*sigmas[0]*sigmas[1]
cov[0,1] = sigma12
cov[1,0] = sigma12
cov
```

```python
# Siccome i pesi del portfolio devono sommare a 1, usiamo la funzione softmax per generare n numeri casuali e farli sommare a 1
def softmax(w):
    a = np.exp(w)
    return a / a.sum()
```

```python
# Ora generiamo i nostri dati. Facciamo 1000 generazioni basandoci sul random walk che conosciamo
# In numpy non dobbiamo fare le trasposizioni con vettori monodimensionali quando usiamo la funzione dot.
N = 1000
returns = np.zeros(N)
risks = np.zeros(N)
for i in range(N):
    w = softmax(np.random.randn(2))
    ret = mean_return.dot(w)
    risk = np.sqrt(w.dot(cov).dot(w)) # Variance = w * cov * w
    returns[i] = ret
    risks[i] = risk
    
```

#### Analisi del plot sotto

Il nostro plot non è un'iperbola come ci immaginavamo. Questo perchè, vedendo i nostri return, abbiamo il primo negativo, ma in valore assoluto molto maggiore rispetto al secondo. Questo significa che qualsiasi portfolio avrà comunque uno sbilanciamento verso ritorni negativi.

Possiamo notare nel plot come gli estremi siano più sbiaditi. Questo perchè la softmax evita di generare valori molto estremi.

Altra cosa che possiamo notare è che il punto in alto del plot rappresenta quando otteniamo il return del secondo asset (investo 100% in R2) mentre il punto in basso rappresenta il return dell'asset negativo, con un investimento del 100% in quell'asset.

```python
# Scatterplottiamo
plt.scatter(risks, returns, alpha=0.1)
plt.xlabel("Risk")
plt.ylabel("Return")
```

```python
# Ora generiamo i pesi in maniera differenti

N = 1000
returns = np.zeros(N)
risks = np.zeros(N)
for i in range(N):
    x = np.random.random()
    w = np.array([x, 1-x])
    ret = mean_return.dot(w)
    risk = np.sqrt(w.dot(cov).dot(w)) # Variance = w * cov * w
    returns[i] = ret
    risks[i] = risk
```

#### Analisi del plot sotto

Questa volta, siccome non usiamo la softmax, distribuiamo uniformemente i pesi e gli estremi sono più scuri.

```python
# Scatterplottiamo
plt.scatter(risks, returns, alpha=0.1)
plt.xlabel("Risk")
plt.ylabel("Return")
```

```python
# Ora generiamo i pesi in maniera differenti, permettiamo i pesi negativi (short-selling)

N = 1000
returns = np.zeros(N)
risks = np.zeros(N)
for i in range(N):
    x = np.random.random() - 0.5 # -0.5 < x < 0.5
    w = np.array([x, 1-x])
    ret = mean_return.dot(w)
    risk = np.sqrt(w.dot(cov).dot(w)) # Variance = w * cov * w
    returns[i] = ret
    risks[i] = risk
```

#### Analisi del plot sotto

Questa volta abbiamo l'iperbole!!
Possiamo notare che questa volta non è simmetrica, perchè siccome x può essere tra -0.5 e 0.5, 1-x sarà tra 0.5 e 1.5!

```python
# Scatterplottiamo
plt.scatter(risks, returns, alpha=0.1)
plt.xlabel("Risk")
plt.ylabel("Return")
```

## 3 Asset Portfolio

```python
# Expected returns da distribuzione normale a media 0 e std dev 0.01
mean_return = np.random.randn(3)*0.01
mean_return
```

```python
# Deviazioni standard. usiamo exponential per renderle positive
sigmas = np.exp(np.random.randn(3))
sigmas
```

```python
# correlation tra TRE variabili. Quindi tra asset 0 e 1, 0 e 2, 1 e 2.
rhos = np.random.randn(3)*0.01
rhos
```

```python
# Ora facciamo un casino per creare la matrice di covarianza.
cov = np.array([
    [sigmas[0]**2, rhos[0]*sigmas[0]*sigmas[1], rhos[1]*sigmas[0]*sigmas[2]],
    [rhos[0]*sigmas[0]*sigmas[1], sigmas[1]**2,  rhos[2]*sigmas[1]*sigmas[2]],
    [rhos[1]*sigmas[0]*sigmas[2], rhos[2]*sigmas[1]*sigmas[2], sigmas[2]**2]
])
```

```python
# Generiamo portfolio con tre assets, 

N = 1000
returns = np.zeros(N)
risks = np.zeros(N)
for i in range(N):
    x1, x2 = np.random.random(2) - 0.5 # -0.5 < x1, x2 < 0.5
    w = np.array([x1, x2, 1-x1-x2])
    np.random.shuffle(w)
    ret = mean_return.dot(w) # Shuffling per bilanciare l'occorrenza del terzo plot
    risk = np.sqrt(w.dot(cov).dot(w)) # Variance = w * cov * w
    returns[i] = ret
    risks[i] = risk
```

#### Analisi del plot sotto

Questa volta non abbiamo una curva ben definita, ma abbiamo una serie di punti al suo interno. Questo succede perchè abbiamo più di 3 asset e non esiste più una formula precisa che lega il risk al return. 

```python
# Scatterplottiamo
plt.scatter(risks, returns, alpha=0.1)
plt.xlabel("Risk")
plt.ylabel("Return")
```
