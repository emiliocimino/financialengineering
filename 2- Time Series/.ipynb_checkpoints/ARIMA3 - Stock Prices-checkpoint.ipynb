# ARIMA on Stock Prices

Applichiamo ARIMA agli stock price. Se gli stock prices sono delle serie temporali, possiamo applicare ARIMA?

```python
def plot_forecast(column, df, train_fit, test_forecast, confint):
    fig, ax = plt.subplots(figsize=(10, 5))

    ax.plot(df.index, df[column], label="True Data")
    ax.plot(train_fit.index, train_fit, label="Train Fit")
    ax.plot(test_forecast.index, test_forecast, label="Forecast")
    ax.fill_between(test_forecast.index, confint[:,0], confint[:,1], color="red", alpha=0.3)
    ax.legend();
```

```python
def rmse(y, t):
    return np.sqrt(np.mean((y-t)**2))
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import pmdarima as pm
```

```python
df = pd.read_csv("sp500sub.csv", index_col="Date", parse_dates=True)
```

```python
df.head()
```

```python
goog = df[df["Name"] == "GOOG"][["Close"]]
goog["Ret"] = goog["Close"].diff()
goog["LogPrice"] = np.log(goog["Close"])
goog["LogRet"] = goog["LogPrice"].diff()
```

```python
goog["LogRet"].plot()
```

```python
N_test = 40
train = goog.iloc[:-N_test]
test = goog.iloc[-N_test:]
```

```python
### ATTENZIONE!!! Se non usi error_action="ignore"
first_model = pm.auto_arima(train["Close"], error_action="ignore", trace=True, seasonal=False, suppress_warnings=True)
```

```python
first_model.summary()
```

```python
first_model.get_params()
```

```python
def predict_and_plot(model, df, column, train, test, N_test):

    params = model.get_params()
    d = params['order'][1]

    train_pred = model.predict_in_sample(start=d, end=-1)
    test_pred, confint = model.predict(n_periods=N_test, return_conf_int=True)

    fig, ax = plt.subplots(figsize=(10, 5))
    ax.plot(df.index, df[column], label="True Data")
    ax.plot(train.index[d:], train_pred, label="Train Fit")
    ax.plot(test.index, test_pred, label="Forecast")
    ax.fill_between(test.index, confint[:,0], confint[:,1], color="red", alpha=0.3)
    ax.set_title("Overall Plot")
    ax.legend();

    
```

```python
def predict_and_plot_test(model, column, test, N_test):

    params = model.get_params()
    d = params['order'][1]

    test_pred, confint = model.predict(n_periods=N_test, return_conf_int=True)

    fig, ax = plt.subplots(figsize=(10, 5))
    ax.plot(test.index, test[column], label="True Data")
    ax.plot(test.index, test_pred, label="Forecast")
    ax.fill_between(test.index, confint[:,0], confint[:,1], color="red", alpha=0.3)
    ax.set_title("Prediction Plot")
    ax.legend();
```

```python
def predict_and_plot_all(model, df, column, train, test, N_test):
    predict_and_plot(model, df, column, train, test, N_test)
    predict_and_plot_test(model, column, test, N_test)
```

```python
predict_and_plot_all(first_model, goog, "Close",  train, test, N_test)
```

```python
print(f"ARIMA RMSE: {rmse(first_model.predict(N_test).to_numpy(), test["Close"])}")
print(f"Naive RMSE: {rmse(train.iloc[-1]["Close"], test["Close"])}")
```

## Conclusioni e Considerazioni:

Abbiamo notato come ARIMA non è effettivamente in grado di calcolare i parametri esatti al fine di battere un Naive Forecast.

ARIMA è in grado di seguire trend e stagionalità, gli stock price pare abbiano un trend debolissimo, e non si conosce niente sulla stagionalità. La supposizione che essi siano un random walk attorno alla media è ancora più valida della possibilità di usare ARIMA. Se dovessimo usare ARIMA sulle stock APPLE, avremmo un problema ancora più grande: ad una certa si verifica un calo drastico che, ovviamente, non può essere predetto da ARIMA.

Un'altra cosa che è possibile fare per esercizio è calcolare la PACF e la ACF degli stock (i log returns, dato che dobbiamo usare serie stazionarie). Calcolando questi valori possiamo notare, in maniera "old school" che i valori p e q per il modello AR e MA sono entrambi 0, indice del fatto che il modello migliore su cui basare le predizioni è esattamente un random walk.

Tuttavia, ci sarà ancora su cui discutere.

```python
second_model = pm.auto_arima(train["Close"], error_action="ignore", trace=True, seasonal=False, suppress_warnings=True,\
                             stepwise=False, max_p=15, max_q=15, max_order=25)
```

```python
predict_and_plot_all(second_model, goog, "Close",  train, test, N_test)
```

```python
print(f"ARIMA RMSE: {rmse(second_model.predict(N_test).to_numpy(), test["Close"])}")
print(f"Naive RMSE: {rmse(train.iloc[-1]["Close"], test["Close"])}")
```
